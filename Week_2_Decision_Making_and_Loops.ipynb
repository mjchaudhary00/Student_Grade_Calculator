Day 2: Repeating Tasks using for loop

```python
for i in range(1, 101):
    print(i)
```

```python
for i in range(1, 51):
    if i % 2 == 0:
        print(i)
```

```python
marks = [95, 82, 67, 45]

for m in marks:
    if m >= 90:
        print(m, "Grade A")
    elif m >= 80:
        print(m, "Grade B")
    elif m >= 70:
        print(m, "Grade C")
    elif m >= 60:
        print(m, "Grade D")
    else:
        print(m, "Grade F")
```

Day 3: Functions in Python

```python
def area_rectangle(length, width):
    return length * width

print(area_rectangle(5, 3))
```

```python
def calculator(a, b, operation):
    if operation == "add":
        return a + b
    elif operation == "sub":
        return a - b
    elif operation == "mul":
        return a * b
    elif operation == "div":
        if b == 0:
            return "Cannot divide by zero"
        return a / b
    else:
        return "Invalid operation"

print(calculator(10, 2, "mul"))
```

Day 4: While Loop & Input Validation

```python
while True:
    marks = int(input("Enter marks (0-100): "))
    if 0 <= marks <= 100:
        break
    else:
        print("Invalid marks. Try again.")
```

Day 5 Project: Student Grade Calculator

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A", "Excellent work! Outstanding performance!"
    elif marks >= 80:
        return "B", "Very Good! Keep it up!"
    elif marks >= 70:
        return "C", "Good effort. Keep practicing!"
    elif marks >= 60:
        return "D", "You passed, but there is room for improvement."
    else:
        return "F", "Don't give up. Learn from mistakes and try again!"
```

```python
# --- Input Section ---
name = input("Enter student name: ").strip().upper()

while True:
    try:
        marks = int(input("Enter marks (0-100): "))
        if 0 <= marks <= 100:
            break
        else:
            print("Invalid marks. Please enter marks between 0 and 100.")
    except ValueError:
        print("Invalid input. Please enter numeric marks only.")


# --- Processing ---
grade, message = calculate_grade(marks)


# --- Output Section ---
print("\nðŸ“Š RESULT FOR {}:".format(name))
print(f"Marks: {marks}/100")
print(f"Grade: {grade}")
print(f"Message: {message}")
```
